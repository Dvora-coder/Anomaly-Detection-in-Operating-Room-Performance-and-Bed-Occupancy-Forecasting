```python
# Loop over each DataFrame in the list dfs
for df in dfs:
    # --- INCISION DATETIME ---
    # Create 'Incision DateTime' if missing
    if 'Incision DateTime' not in df.columns:
        if 'Incision Date' in df.columns and 'Incision Time' in df.columns:
            df['Incision DateTime'] = pd.to_datetime(
                df['Incision Date'].astype(str) + ' ' + df['Incision Time'].astype(str),
                errors='coerce'
            )
        # If there is only incision time (assume same date as entry to OR)
        elif 'Incision Time' in df.columns and 'Actual Surgery Room Entry Date' in df.columns:
            df['Incision DateTime'] = pd.to_datetime(
                df['Actual Surgery Room Entry Date'].astype(str) + ' ' + df['Incision Time'].astype(str),
                errors='coerce'
            )
        # entry date + entry time (actual time patient enters the OR)
        elif 'Actual Surgery Room Entry Date' in df.columns and 'Actual Surgery Room Entry Time' in df.columns:
            df['Incision DateTime'] = pd.to_datetime(
                df['Actual Surgery Room Entry Date'].astype(str) + ' ' + df['Actual Surgery Room Entry Time'].astype(str),
                errors='coerce'
            )
        else:
            df['Incision DateTime'] = pd.NaT


    # --- CLOSURE DATETIME ---
    # Create 'Closure DateTime' if missing
    if 'Closure DateTime' not in df.columns:
        if 'Closure Date' in df.columns and 'Closure Time' in df.columns:
            df['Closure DateTime'] = pd.to_datetime(
                df['Closure Date'].astype(str) + ' ' + df['Closure Time'].astype(str),
                errors='coerce'
            )
        # If only closure time exists, assume same date as entry
        elif 'Closure Time' in df.columns and 'Actual Surgery Room Entry Date' in df.columns:
            df['Closure DateTime'] = pd.to_datetime(
                df['Actual Surgery Room Entry Date'].astype(str) + ' ' + df['Closure Time'].astype(str),
                errors='coerce'
            )
        else:
            df['Closure DateTime'] = pd.NaT
```

```python
# Loop over each DataFrame in the list dfs
for df in dfs:
    # --- ENTRY DATETIME ---
    # Create 'Entry DateTime' if missing, by combining date and time columns
    if 'Entry DateTime' not in df.columns:
        # If both separate date and time columns exist
        if 'Actual Surgery Room Entry Date' in df.columns and 'Actual Surgery Room Entry Time' in df.columns:
            df['Entry DateTime'] = pd.to_datetime(
                df['Actual Surgery Room Entry Date'].astype(str) + ' ' + df['Actual Surgery Room Entry Time'].astype(str),
                errors='coerce'
            )
        else:
            # If missing necessary columns, set as NaT
            df['Entry DateTime'] = pd.NaT

    # --- PLANNED START DATETIME ---
    # Create 'Planned Start DateTime' if missing
    if 'Planned Start DateTime' not in df.columns:
        if 'Planned Surgery Date' in df.columns and 'Planned Surgery Time' in df.columns:
            df['Planned Start DateTime'] = pd.to_datetime(
                df['Planned Surgery Date'].astype(str) + ' ' + df['Planned Surgery Time'].astype(str),
                errors='coerce'
            )
        else:
            df['Planned Start DateTime'] = pd.NaT

    # --- EXIT DATETIME ---
    # Create 'Exit DateTime' if missing
    if 'Exit DateTime' not in df.columns:
        # Prefer combining 'End of Surgery Date (Exit from OR)' and 'End of Surgery Time (Exit from OR)'
        if 'End of Surgery Date (Exit from OR)' in df.columns and 'End of Surgery Time (Exit from OR)' in df.columns:
            df['Exit DateTime'] = pd.to_datetime(
                df['End of Surgery Date (Exit from OR)'].astype(str) + ' ' + df['End of Surgery Time (Exit from OR)'].astype(str),
                errors='coerce'
            )

        elif 'Closure DateTime' in df.columns:
            df['Exit DateTime'] = df['Closure DateTime']
        else:
            df['Exit DateTime'] = pd.NaT
```

In this step, we prepare the dataset for machine learning by removing potential data leakage columns — features that would not be available at prediction time (such as actual surgery or discharge times). This ensures the integrity of the model evaluation. In the next steps, we will create meaningful new features, and structure the data appropriately for modeling.

#### Predictive Modeling for Surgery Duration

In this section, we build and compare several machine learning models to predict the actual duration of surgeries. The process begins by filtering out invalid target values and selecting only numeric features. A Random Forest model is used for feature importance ranking to reduce dimensionality and retain only relevant predictors.

The target variable (surgery duration) is transformed using a log scale (`log1p`) to reduce skewness and improve model performance. Multiple regression models are then evaluated using 5-fold cross-validation, including Random Forest, XGBoost, LightGBM, CatBoost, and ElasticNet. The performance of each model is assessed using R² and Mean Absolute Error (MAE).

Finally, the best-performing model is selected based on cross-validation results and is evaluated on a hold-out test set. Predictions are converted back to the original scale using the inverse transformation for interpretation in minutes.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import make_scorer, r2_score, mean_absolute_error
from sklearn.linear_model import ElasticNet
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor

# Filtering rows with valid values in the target variable
valid_rows = df_copy['actual_surgery_duration'].notna() & (df_copy['actual_surgery_duration'] > 0)
df_valid = df_copy.loc[valid_rows].copy()

# Split into X and Y
X = df_valid.drop(columns=['actual_surgery_duration'])
y = df_valid['actual_surgery_duration']

# Keep only numeric columns + remove rows with missing values
X_num = X.select_dtypes(include=[np.number]).dropna()
y = y.loc[X_num.index]

# Feature selection with RandomForest
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_num, y)

importances = pd.Series(rf.feature_importances_, index=X_num.columns)
important_features = importances[importances > 0.01].sort_values(ascending=False)
X_selected = X_num[important_features.index]
y = y.loc[X_selected.index]

# The conversion to log1p
y_log = np.log1p(y)

# Split into Train/Test
X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y_log, test_size=0.2, random_state=42
)

scoring = {
    'R2':  make_scorer(r2_score),
    'MAE': make_scorer(mean_absolute_error, greater_is_better=False)
}

# Comparison Models
models = {
    'RandomForest': RandomForestRegressor(random_state=42),
    'XGBoost':       XGBRegressor(random_state=42, tree_method='hist', eval_metric='rmse'),
    'LightGBM':      LGBMRegressor(random_state=42),
    'CatBoost':      CatBoostRegressor(random_state=42, verbose=0),
    'ElasticNet':    ElasticNet(random_state=42)
}

# Basic pipeline
base_pipeline = Pipeline([
    ('imputer',  SimpleImputer(strategy='mean')),
    ('scaler',   StandardScaler())
])

# Training and performance evaluation with cross_validate
results = []
for name, model in models.items():
    pipe = Pipeline([
        *base_pipeline.steps,
        ('regressor', model)
    ])
    cv = cross_validate(
        pipe, X_train, y_train,
        cv=5, scoring=scoring,
        return_train_score=False
    )
    results.append({
        'model': name,
        'R2_mean':  cv['test_R2'].mean(),
        'R2_std':   cv['test_R2'].std(),
        'MAE_mean': -cv['test_MAE'].mean(),
        'MAE_std':  cv['test_MAE'].std()
    })

# Comparison table of results
df_results = pd.DataFrame(results).set_index('model').sort_values('R2_mean', ascending=False)
print("\n=== Model Comparison ===")
print(df_results)

# Evaluation of the best model on the test set
best_model_name = df_results.index[0]
best_model = models[best_model_name]

best_pipe = Pipeline([
    *base_pipeline.steps,
    ('regressor', best_model)
])
best_pipe.fit(X_train, y_train)
y_pred_log = best_pipe.predict(X_test)

y_pred = np.expm1(y_pred_log)
y_pred[y_pred < 1] = 1

print(f"\n=== Hold-out Results for {best_model_name} ===")
print("R² (log scale):", r2_score(y_test, y_pred_log))
print("R² (original scale):", r2_score(np.expm1(y_test), y_pred))
print("MAE (minutes):", mean_absolute_error(np.expm1(y_test), y_pred))
```

```python
def merge_intervals(intervals):
    """
    Receives a list of tuples (start, end) and returns a list of merged intervals.
    """
    intervals = sorted(intervals, key=lambda x: x[0])
    mergedlist = []
    for s, e in intervals:
        if not mergedlist or s > mergedlist[-1][1]:
            mergedlist.append([s, e])
        else:
            mergedlist[-1][1] = max(mergedlist[-1][1], e)
    return [(s, e) for s, e in mergedlist]
```

### Schedule Validation: Overlap and Duplication Checks

This function validates the integrity of the surgical scheduling DataFrame by checking for potential conflicts. It detects overlapping time slots for each entity involved in the schedule — including operating rooms, surgeons, anesthesiologists, and surgical team members — based on the `start_time`, `end_time`, and `day_of_week` fields.

The function also checks for duplicate `case_id`s, which must be unique. Overlaps are identified when one case starts before the previous one ends. Surgical teams are parsed into individual members to ensure no team member is scheduled for two overlapping surgeries. The output is a list of descriptive error messages that indicate the type and location of each detected issue. If no conflicts are found, a confirmation message is printed.

```python
import pandas as pd

def check_schedule_duplicates(schedule_df):
    """
    Checks for overlaps and duplicate case_ids in a surgical scheduling DataFrame.
    The DataFrame must have the following columns:
    - 'case_id': Unique identifier for each case
    - 'room': Operating room number
    - 'surgeon': Surgeon ID
    - 'anes': Anesthesiologist ID
    - 'team': Team members (comma-separated string or list)
    - 'start_time': Surgery start time (numeric or datetime)
    - 'end_time': Surgery end time (numeric or datetime)
    - 'day_of_week' or 'date': If scheduling covers multiple days

    Returns:
        errors: List of strings describing detected overlaps and duplicates.
    """
    errors = []

    # Checks for time overlaps for a given entity (room, surgeon, etc.)
    def check_entity_overlap(df, entity_col, entity_id, desc):
        df = df.sort_values('start_time')
        for i in range(len(df) - 1):
            current_end = df.iloc[i]['end_time']
            next_start = df.iloc[i+1]['start_time']
            # Overlap detected: next surgery starts before current one ends
            if next_start < current_end:
                errors.append(
                    f"Overlap {desc}: {entity_col} {entity_id} - case_ids {df.iloc[i]['case_id']} & {df.iloc[i+1]['case_id']}"
                )

    # Check for overlaps in each room on each day
    for (room, day), group in schedule_df.groupby(['room', 'day_of_week']):
        check_entity_overlap(group, 'room', room, f"room {room} on day {day}")

    # Check for overlaps for each surgeon on each day
    for (surgeon, day), group in schedule_df.groupby(['surgeon', 'day_of_week']):
        check_entity_overlap(group, 'surgeon', surgeon, f"surgeon {surgeon} on day {day}")

    # Check for overlaps for each anesthesiologist on each day
    for (anes, day), group in schedule_df.groupby(['anes', 'day_of_week']):
        check_entity_overlap(group, 'anes', anes, f"anes {anes} on day {day}")

    # Check for overlaps for each team member on each day
    def parse_team(x):
        if isinstance(x, list):
            return x
        if pd.isnull(x):
            return []
        if isinstance(x, str):
            return [int(xx.strip()) for xx in x.split(',') if xx.strip().isdigit()]
        return []
    schedule_df['team_list'] = schedule_df['team'].apply(parse_team)
    for day, group in schedule_df.groupby('day_of_week'):
        all_members = set(m for members in group['team_list'] for m in members)
        for member in all_members:
            member_cases = group[group['team_list'].apply(lambda l: member in l)]
            check_entity_overlap(member_cases, 'team', member, f"team member {member} on day {day}")

    # 5Check for duplicate case_ids (should be unique)
    duplicates = schedule_df['case_id'][schedule_df['case_id'].duplicated()]
    if not duplicates.empty:
        errors.append(f"Duplicate case_ids found: {duplicates.unique()}")

    if not errors:
        print("No overlaps or duplicate case_ids found!")
    else:
        print("\n".join(errors))
    return errors
```

### Calculating Room Utilization Based on Surgeon Availability

This section defines a utility function (`merge_intervals`) and a main function (`calc_room_utilization_surgeon_blocks`) that together calculate the utilization of each operating room per day, based on the availability blocks of the surgeons assigned to that room.

Instead of assuming full-day room availability, the approach aggregates and merges the actual availability intervals of the surgeons. It then calculates the total scheduled surgery time and divides it by the total available time to obtain a realistic utilization ratio.

This method reflects a more accurate picture of room efficiency by incorporating real-world constraints such as surgeon-specific availability. The output is a summary DataFrame with the room number, day of week, total scheduled time, surgeon-available time, utilization ratio, and the list of case IDs included.

```python
# Utility function to merge overlapping intervals (used for surgeon blocks)
def merge_intervals(intervals):
    """
    Given a list of intervals (start, end), merge overlapping intervals into a minimal set.
    Returns a list of merged, non-overlapping intervals.
    """
    if not intervals:
        return []
    sorted_int = sorted(intervals)
    merged = [sorted_int[0]]
    for current in sorted_int[1:]:
        last = merged[-1]
        # If intervals overlap, merge them
        if current[0] <= last[1]:
            merged[-1] = (last[0], max(last[1], current[1]))
        else:
            merged.append(current)
    return merged

# Main function to calculate room utilization based on surgeon availability blocks
def calc_room_utilization_surgeon_blocks(schedule_df, avail_surgeon):
    """
    For each room and day, calculate the utilization as the sum of all scheduled cases
    divided by the total available time (as defined by the merged blocks of all surgeons assigned to that room).
    Returns a DataFrame with room, day, total used time, available time, utilization ratio, and case IDs.
    """
    utilization = []
    for room in schedule_df['room'].unique():
        for day in schedule_df[schedule_df['room'] == room]['day_of_week'].unique():
            day_df = schedule_df[(schedule_df['room'] == room) & (schedule_df['day_of_week'] == day)]
            surgeons_in_room = day_df['surgeon'].unique()
            surgeon_blocks = []
            # Collect all availability blocks for surgeons assigned to this room and day
            for surgeon in surgeons_in_room:
                if surgeon in avail_surgeon and day in avail_surgeon[surgeon]:
                    surgeon_blocks.extend(avail_surgeon[surgeon][day])
            # Merge overlapping blocks for all surgeons in this room/day
            merged_blocks = merge_intervals(surgeon_blocks)
            total_surgeon_available = sum(e-s for s, e in merged_blocks)
            # Calculate total scheduled case time in this room/day
            total_case_time = (day_df['end_time'] - day_df['start_time']).sum()
            utilization.append({
                'room': room,
                'day_of_week': day,
                'used': total_case_time,
                'available': total_surgeon_available,
                'utilization': total_case_time / total_surgeon_available if total_surgeon_available > 0 else 0,
                'in_block_cases': list(day_df['case_id'])
            })
    return pd.DataFrame(utilization)
```

### Weekly Surgery Scheduling with CP-SAT Optimization

This section implements the core constraint programming model for scheduling surgeries over a weekly horizon using Google OR-Tools’ CP-SAT solver.

The `solve_week()` function receives a list of surgical cases for the week and performs the following steps:

- **Decision Variables**: For each case, day, and room (if allowed), it creates binary assignment variables and corresponding integer variables for start times.
- **Assignment Constraints**: Each surgery can be scheduled at most once (one room and one day).
- **Availability Constraints**: Surgeries can only be assigned within the available time blocks of the operating room, surgeon, and anesthesiologist.
- **No-Overlap Constraints**: Ensures that no two surgeries overlap for the same room, surgeon, or anesthesiologist on the same day. These are modeled using optional interval variables.
- **Objective Function**: The goal is to maximize total scheduled surgery time while minimizing the number of rooms opened (via penalty terms).
- **Output**: Returns a DataFrame of scheduled surgeries including room, day, start and end times, and involved personnel.

In addition, the code defines:
- `merge_intervals()`: A utility to merge overlapping time blocks.
- `calc_room_day_available_time()`: Calculates actual available minutes in each room per day based on the surgeons assigned.
- `calc_room_day_utilization()`: Computes the room utilization ratio (used time vs. available time), which reflects real efficiency based on surgeon-specific availability rather than theoretical room hours.


```python
from ortools.sat.python import cp_model

def solve_week(
    week_cases_df,
    avail_or,
    avail_surgeon,
    avail_anes=None,
    pref_sr=None,
    pref_sa=None,
    pref_st=None,
    buffer=10,
    penalty_open_room=1
):
    # Columns for the output DataFrame if there are no results
    EMPTY_COLUMNS = ['day_of_week', 'case_id', 'room', 'start_time', 'end_time', 'surgeon', 'anes', 'team']

    df = week_cases_df.copy()
    days = sorted(df['day_of_week'].unique())
    rooms = list(avail_or.keys())
    cases = df['case_id'].tolist()
    surgeons = df.set_index('case_id')['surgeon'].to_dict()
    anesthes = df.set_index('case_id')['anes'].to_dict()

    # Parse team codes into lists
    def parse_team(x):
        if isinstance(x, list):
            return x
        if pd.isnull(x):
            return []
        if isinstance(x, str):
            return [int(xx.strip()) for xx in x.split(',') if xx.strip().isdigit()]
        return []
    df['team_all'] = df['team_all'].apply(parse_team)
    team_dict = df.set_index('case_id')['team_all'].to_dict()

    durations = df.set_index('case_id')['pred_duration'].to_dict()
    room_options_dict = df.set_index('case_id')['room_options'].to_dict()  # each case has its allowed rooms

    model = cp_model.CpModel()
    assign = {}
    start_time = {}

    # Decision variables: only for allowed rooms per case
    for c in cases:
        allowed_rooms = room_options_dict.get(c, [])
        for d in days:
            for r in rooms:
                if r in allowed_rooms:
                    assign[(c, r, d)] = model.NewBoolVar(f'assign_{c}_{r}_{d}')
                    start_time[(c, r, d)] = model.NewIntVar(0, 50000, f'start_{c}_{r}_{d}')
                else:
                    assign[(c, r, d)] = model.NewConstant(0)

    # Each case can be scheduled at most once in the week (one room and one day)
    for c in cases:
        allowed_rooms = room_options_dict.get(c, [])
        model.Add(
            sum(assign[(c, r, d)] for r in rooms for d in days if r in allowed_rooms) <= 1
        )

    # Room and surgeon availability constraints (could add more for anesthesiologist/team)
    bigM = 1000000
    for c in cases:
        dur = int(durations[c])
        s = surgeons[c]
        for d in days:
            for r in rooms:
                if r not in room_options_dict.get(c, []):
                    continue
                a = anesthes[c]
                if (
                    (d not in avail_or[r])
                    or (d not in avail_surgeon.get(s, {}))
                    or (avail_anes is not None and (a not in avail_anes or d not in avail_anes[a]))
                ):
                    model.Add(assign[(c, r, d)] == 0)
                    continue

                slots_room = avail_or[r][d]
                slots_surg = avail_surgeon[s][d]
                slots_anes = avail_anes[a][d] if avail_anes and a in avail_anes and d in avail_anes[a] else []
                allowed = []
                for (start_r, end_r) in slots_room:
                    for (start_s, end_s) in slots_surg:
                        for (start_a, end_a) in slots_anes:
                            start = int(max(start_r, start_s, start_a))
                            end = int(min(end_r, end_s, end_a))
                            if end - start >= dur:
                                allowed.append((start, end))
                if allowed:
                    min_start = min(a[0] for a in allowed)
                    max_end = max(a[1] for a in allowed)
                    # If assigned, must start within the available block
                    model.Add(start_time[(c, r, d)] >= min_start - (1 - assign[(c, r, d)]) * bigM)
                    model.Add(start_time[(c, r, d)] + dur <= max_end + (1 - assign[(c, r, d)]) * bigM)
                else:
                    model.Add(assign[(c, r, d)] == 0)

    # No-overlap constraints for rooms, surgeons, and anesthesiologists
    # For each room on each day, no overlap between surgeries
    for r in rooms:
        for d in days:
            rcases = [c for c in cases if r in room_options_dict.get(c, [])]
            intervals = []
            for c in rcases:
                dur = int(durations[c]) + buffer
                intervals.append(
                    model.NewOptionalIntervalVar(
                        start_time[(c, r, d)], dur,
                        start_time[(c, r, d)] + dur,
                        assign[(c, r, d)], f'int_room_{c}_{r}_{d}'
                    )
                )
            if intervals:
                model.AddNoOverlap(intervals)

    # For each surgeon on each day, no overlap between their surgeries
    for s in set(surgeons.values()):
        for d in days:
            scases = [c for c in cases if surgeons[c] == s]
            intervals = []
            for c in scases:
                for r in rooms:
                    if r in room_options_dict.get(c, []):
                        dur = int(durations[c]) + buffer
                        intervals.append(
                            model.NewOptionalIntervalVar(
                                start_time[(c, r, d)], dur,
                                start_time[(c, r, d)] + dur,
                                assign[(c, r, d)], f'int_surg_{c}_{r}_{d}_{s}'
                            )
                        )
            if intervals:
                model.AddNoOverlap(intervals)

    # For each anesthesiologist, prevent overlaps if required (similar loop)
    if avail_anes is not None:
        for a in set(anesthes.values()):
            for d in days:
                acases = [c for c in cases if anesthes[c] == a]
                intervals = []
                for c in acases:
                    for r in rooms:
                        if r in room_options_dict.get(c, []):
                            dur = int(durations[c]) + buffer
                            intervals.append(
                                model.NewOptionalIntervalVar(
                                    start_time[(c, r, d)], dur,
                                    start_time[(c, r, d)] + dur,
                                    assign[(c, r, d)], f'int_anes_{c}_{r}_{d}_{a}'
                                )
                            )
                if intervals:
                    model.AddNoOverlap(intervals)

    # Maximize total scheduled time minus penalty for each open room
    room_used = {}
    for r in rooms:
        for d in days:
            assign_list = [assign[(c, r, d)] for c in cases if r in room_options_dict.get(c, [])]
            if assign_list:
                room_used[(r, d)] = model.NewBoolVar(f'room_used_{r}_{d}')
                model.AddMaxEquality(room_used[(r, d)], assign_list)
            else:
                room_used[(r, d)] = model.NewConstant(0)

    # Maximize total scheduled surgery time, penalize opening many rooms
    terms = []
    for c in cases:
        for r in rooms:
            for d in days:
                if r in room_options_dict.get(c, []):
                    terms.append(assign[(c, r, d)] * durations[c])
    model.Maximize(
        sum(terms)
        - penalty_open_room * sum(room_used[(r, d)] for r in rooms for d in days)
    )

    # Solve the model
    solver = cp_model.CpSolver()
    solver.parameters.max_time_in_seconds = 320
    status = solver.Solve(model)

    # Collect results
    scheduled = []
    for c in cases:
        for r in rooms:
            for d in days:
                if r in room_options_dict.get(c, []):
                    if solver.Value(assign[(c, r, d)]):
                        start = solver.Value(start_time[(c, r, d)])
                        scheduled.append({
                            'day_of_week': d,
                            'case_id': c,
                            'room': r,
                            'start_time': start,
                            'end_time': start + int(durations[c]),
                            'surgeon': surgeons[c],
                            'anes': anesthes[c],
                            'team': ','.join(str(t) for t in team_dict[c])
                        })
    result_df = pd.DataFrame(scheduled)
    if not result_df.empty and 'start_time' in result_df.columns:
        result_df = result_df.sort_values(['day_of_week', 'room', 'start_time'])
    else:
        result_df = pd.DataFrame([], columns=EMPTY_COLUMNS)
    print(f"\n=== Week Schedule ({len(result_df)} cases scheduled) ===")
    print(result_df)
    return result_df

# Merge overlapping intervals
def merge_intervals(intervals):
    """
    Takes a list of intervals [(start, end), ...] and merges overlapping intervals into a list with no overlaps.
    """
    if not intervals:
        return []
    sorted_int = sorted(intervals)
    merged = [sorted_int[0]]
    for current in sorted_int[1:]:
        last = merged[-1]
        if current[0] <= last[1]:
            merged[-1] = (last[0], max(last[1], current[1]))
        else:
            merged.append(current)
    return merged

def calc_room_day_available_time(result_df, avail_surgeon):
    """
    For each room+day, finds all surgeons scheduled in that room on that day,
    merges all their available blocks and returns the total real available minutes.
    """
    output = []
    for (room, day), group in result_df.groupby(['room', 'day_of_week']):
        surgeons = group['surgeon'].unique()
        intervals = []
        for s in surgeons:
            if day in avail_surgeon[s]:
                intervals.extend(avail_surgeon[s][day])
        merged = merge_intervals(intervals)
        total_available = sum(end - start for start, end in merged)
        output.append({'room': room, 'day_of_week': day, 'available': total_available})
    return pd.DataFrame(output)

def calc_room_day_utilization(result_df, avail_surgeon):
    """
    Calculates the utilization rate for each room+day.
    Utilization = total scheduled time / total available time.
    """
    room_day_avail = calc_room_day_available_time(result_df, avail_surgeon)
    room_day_used = (
        result_df
        .assign(duration=lambda x: x['end_time'] - x['start_time'])
        .groupby(['room', 'day_of_week'])['duration']
        .sum()
        .reset_index()
        .rename(columns={'duration': 'used'})
    )
    util_df = pd.merge(room_day_used, room_day_avail, on=['room', 'day_of_week'], how='left')
    util_df['utilization'] = util_df['used'] / util_df['available']
    return util_df
```

**Greedy Scheduling Algorithm with Utilization Threshold**

This function implements a greedy heuristic to schedule surgeries into operating rooms
based on available time slots and resource constraints. It iterates through surgical cases sorted by descending predicted duration and tries to assign each case to a feasible time window while respecting availability of:

- Operating rooms
- Surgeons
- Anesthesiologists
- Surgical team members

Key features:
- Enforces buffer time between consecutive surgeries
- Prioritizes rooms with utilization below a defined threshold (`util_thresh`)
- Avoids scheduling conflicts by checking for overlaps across all involved entities
- Updates availability dynamically after each assignment
- Returns two DataFrames: one with successfully scheduled cases, and one with leftovers

This approach is suitable for quick initial scheduling before applying a more sophisticated
optimization model.

```python
import copy

# Removes a time slot from the list of available slots, adjusting remaining availability
def remove_time_slot(slots, remove_start, remove_end):
    new_slots = []
    for (s, e) in slots:
        if e <= remove_start or s >= remove_end:
            new_slots.append((s, e))
        else:
            if s < remove_start:
                new_slots.append((s, remove_start))
            if remove_end < e:
                new_slots.append((remove_end, e))
    return new_slots

# Greedy scheduling function that tries to assign surgeries to rooms with utilization threshold
def greedy_fill_rooms_with_utilization_threshold(
    new_cases_df,
    avail_or,
    avail_surgeon,
    avail_anes=None,
    avail_team=None,
    pref_sr=None,
    buffer=10,
    util_thresh=0.5
):
    # Deep copy of availability dictionaries to avoid modifying the originals
    or_avail = {r: {d: list(avail_or[r][d]) for d in avail_or[r]} for r in avail_or}

    # Important: loop through the keys of avail_surgeon, not avail_surgeon[s]
    surg_avail = {s: { d: list(avail_surgeon[s][d]) for d in avail_surgeon[s] } for s in avail_surgeon}
    anes_avail = {a: {d: list(avail_anes[a][d]) for d in avail_anes[a]} for a in avail_anes} if avail_anes else None
    team_avail = {m: {d: list(avail_team[m][d]) for d in avail_team[m]} for m in avail_team} if avail_team else None

    # Check for overlap between a new slot and any existing slots
    def conflict(slot, slots):
        s0, e0 = slot
        return any(not (e0 <= s or s0 >= e) for (s, e) in slots)

    # Initialize room availability and usage tracking
    room_available = {(r, d): sum(e - s for (s, e) in or_avail[r][d]) for r in or_avail for d in or_avail[r]}
    room_used_time = {key: 0 for key in room_available}

    # Tracking used time slots for various resources
    used_room = {}
    used_surg = {}
    used_anes = {}
    used_team = {}
    scheduled = []
    leftovers = []

    # Sort cases by predicted duration in descending order
    cases = new_cases_df.sort_values('pred_duration', ascending=False).to_dict('records')

    for case in cases:
        cid = case['case_id']
        day = case['day_of_week']
        dur = int(case['pred_duration'])
        surg = case['surgeon']
        anes = case.get('anes')
        team = []

        # Extract team member list
        if 'team_all' in case and isinstance(case['team_all'], list):
            team = case['team_all']
        elif 'team' in case:
            if isinstance(case['team'], list): team = case['team']
            elif isinstance(case['team'], str):
                team = [int(x.strip()) for x in case['team'].split(',') if x.strip().isdigit()]

        # Check if surgeon is available on this day
        if day not in surg_avail.get(surg, {}):
            leftovers.append(cid)
            continue

        # Filter operating rooms available on this day
        rooms = [r for r in or_avail if day in or_avail[r]]
        if anes and anes_avail:
            rooms = [r for r in rooms if day in anes_avail.get(anes, {})]
        if team and team_avail:
            rooms = [r for r in rooms if all(day in team_avail.get(m, {}) for m in team)]

        if not rooms:
            leftovers.append(cid)
            continue

        # Prioritize rooms with utilization below threshold
        below = [r for r in rooms if room_used_time[(r, day)] / max(room_available[(r, day)], 1) < util_thresh]
        ordered = below + [r for r in rooms if r not in below]

        placed = False
        for r in ordered:
            for (rs, re) in or_avail[r][day]:
                for (ss, se) in surg_avail[surg][day]:
                    start = max(rs, ss)
                    end = start + dur
                    if end > min(re, se): continue

                    # Check anesthesiologist availability
                    if anes and anes_avail:
                        if not any(start >= as0 and end <= ae0 for (as0, ae0) in anes_avail[anes][day]):
                            continue
                    # Check team availability
                    if team and team_avail:
                        if not all(any(start >= ts0 and end <= te0 for (ts0, te0) in team_avail[m][day]) for m in team):
                            continue

                    # Check for conflicts with already scheduled slots
                    slot = (start, end + buffer)
                    if (conflict(slot, used_room.get((r, day), [])) or
                        conflict(slot, used_surg.get((surg, day), [])) or
                        (anes and conflict(slot, used_anes.get((anes, day), []))) or
                        any(conflict(slot, used_team.get((m, day), [])) for m in team)):
                        continue

                    # Assign surgery to schedule
                    scheduled.append({
                        'case_id': cid,
                        'day_of_week': day,
                        'room': r,
                        'surgeon': surg,
                        'start_time': start,
                        'end_time': end,
                        'anes': anes,
                        'team_all': team
                    })

                    # Update usage tracking
                    used_room.setdefault((r, day), []).append(slot)
                    used_surg.setdefault((surg, day), []).append(slot)
                    if anes: used_anes.setdefault((anes, day), []).append(slot)
                    for m in team:
                        used_team.setdefault((m, day), []).append(slot)

                    # Remove assigned slot from available time
                    or_avail[r][day] = remove_time_slot(or_avail[r][day], slot[0], slot[1])
                    surg_avail[surg][day] = remove_time_slot(surg_avail[surg][day], slot[0], slot[1])
                    if anes: anes_avail[anes][day] = remove_time_slot(anes_avail[anes][day], slot[0], slot[1])
                    for m in team:
                        team_avail[m][day] = remove_time_slot(team_avail[m][day], slot[0], slot[1])

                    room_used_time[(r, day)] += dur
                    placed = True
                    break
                if placed: break
            if placed: break

        # If not placed in any room, mark as leftover
        if not placed:
            leftovers.append(cid)

    # Return scheduled cases and leftover cases as DataFrames
    scheduled_df = pd.DataFrame(scheduled)
    leftovers_df = new_cases_df[new_cases_df['case_id'].isin(leftovers)]
    return scheduled_df, leftovers_df
```

### Multi-Stage Surgery Scheduling Framework

This script implements a complete end-to-end weekly surgery scheduling pipeline that combines:
1. **Greedy Scheduling** — fast allocation of cases to underutilized operating rooms while satisfying surgeon, anesthesiologist, and team availability.
2. **Constraint Programming (CP)** — used to schedule remaining "leftover" cases with more complex constraints and maximize resource utilization.
3. **Overlap Resolution & Iterative Rescheduling** — detects and resolves conflicts (e.g., double-booked resources) by removing conflicting assignments and re-running CP or greedy heuristics as needed.
4. **Utilization Analysis** — calculates the real utilization rate per room and day based on the total scheduled time vs. available time derived from surgeon availability.
5. **Output Generation** — saves final scheduling and utilization results to CSV files.

This framework balances speed and feasibility (via greedy assignment) with optimality and fairness (via constraint-based scheduling), making it suitable for real-world hospital operating room management.


```python
import time
import copy

buffer = 10


def remove_time_slot(slots, remove_start, remove_end):
    new_slots = []
    for (start, end) in slots:
        if end <= remove_start or start >= remove_end:
            new_slots.append((start, end))
        else:
            if start < remove_start:
                new_slots.append((start, remove_start))
            if remove_end < end:
                new_slots.append((remove_end, end))
    return new_slots

def merge_intervals(intervals):
    if not intervals:
        return []
    sorted_int = sorted(intervals)
    merged = [sorted_int[0]]
    for current in sorted_int[1:]:
        last = merged[-1]
        if current[0] <= last[1]:
            merged[-1] = (last[0], max(last[1], current[1]))
        else:
            merged.append(current)
    return merged

def check_schedule_duplicates(schedule_df):
    errors = []

    def check_entity_overlap(df, entity_col, entity_id, desc):
        df = df.sort_values('start_time')
        for i in range(len(df) - 1):
            id1 = df.iloc[i]['case_id']
            id2 = df.iloc[i+1]['case_id']
            if id1 == id2:
                continue
            current_end = df.iloc[i]['end_time']
            next_start = df.iloc[i+1]['start_time']
            if next_start < current_end:
                errors.append(f"Overlap {desc}: {entity_col} {entity_id} - case_ids {id1} & {id2}")

    # room overlaps
    for (room, day), group in schedule_df.groupby(['room', 'day_of_week']):
        check_entity_overlap(group, 'room', room, f"room {room} on day {day}")
    # surgeon overlaps
    for (surgeon, day), group in schedule_df.groupby(['surgeon', 'day_of_week']):
        check_entity_overlap(group, 'surgeon', surgeon, f"surgeon {surgeon} on day {day}")
    # anesthes overlaps
    for (anes, day), group in schedule_df.groupby(['anes', 'day_of_week']):
        check_entity_overlap(group, 'anes', anes, f"anes {anes} on day {day}")
    # team overlaps
    def parse_team(x):
        if isinstance(x, list): return x
        if pd.isnull(x): return []
        if isinstance(x, str):
            return [int(xx.strip()) for xx in x.split(',') if xx.strip().isdigit()]
        return []
    schedule_df = schedule_df.copy()
    schedule_df['team_list'] = schedule_df['team'].apply(parse_team)
    for day, group in schedule_df.groupby('day_of_week'):
        members = set(m for sub in group['team_list'] for m in sub)
        for m in members:
            cases_m = group[group['team_list'].apply(lambda l: m in l)]
            check_entity_overlap(cases_m, 'team', m, f"team member {m} on day {day}")
    # duplicate case_ids
    dupe_ids = schedule_df['case_id'][schedule_df['case_id'].duplicated()]
    if not dupe_ids.empty:
        errors.append(f"Duplicate case_ids found: {dupe_ids.unique()}")

    if not errors:
        print("No overlaps or duplicate case_ids found!")
    else:
        print("\n".join(errors))
    return errors

# --- Release resources for removed case ---
def release_case_slots(case_row, avail_or, avail_surgeon, avail_anes, avail_team, buffer=10):
    room = case_row['room']
    day = case_row['day_of_week']
    surgeon = case_row['surgeon']
    anes = case_row.get('anes', None)
    raw_team = case_row.get('team_all', case_row.get('team', [])) or []
    team = []
    if isinstance(raw_team, str):
        for xx in raw_team.split(','):
            if xx.strip().isdigit():
                team.append(int(xx.strip()))
    elif isinstance(raw_team, list):
        team = [int(t) for t in raw_team if pd.notnull(t)]
    s, e = case_row['start_time'], case_row['end_time']
    rs, re = s, e + buffer
    if room in avail_or and day in avail_or[room]:
        avail_or[room][day].append((rs, re))
        avail_or[room][day] = merge_intervals(avail_or[room][day])
    if surgeon in avail_surgeon and day in avail_surgeon[surgeon]:
        avail_surgeon[surgeon][day].append((rs, re))
        avail_surgeon[surgeon][day] = merge_intervals(avail_surgeon[surgeon][day])
    if anes is not None and anes in avail_anes and day in avail_anes[anes]:
        avail_anes[anes][day].append((rs, re))
        avail_anes[anes][day] = merge_intervals(avail_anes[anes][day])
    for m in team:
        if m in avail_team and day in avail_team[m]:
            avail_team[m][day].append((rs, re))
            avail_team[m][day] = merge_intervals(avail_team[m][day])

# --- Overlap-resolution & rerun CP ---
def resolve_overlaps_and_rerun_cp(
    final_schedule, cp_func,
    avail_or_week, avail_surgeon_week,
    avail_anes_week, avail_team_week,
    pref_sr, pref_sa, pref_st,
    buffer, original_cases_df,
    greedy_first=True, max_total_iterations=25
):
    rerun_counter = 0
    prev_num_errors = -1
    while True:
        overlaps = check_schedule_duplicates(final_schedule)
        num_errors = len(overlaps)
        if num_errors == 0:
            print(f"No overlaps after {rerun_counter} reruns; total {len(final_schedule)} cases")
            break
        if num_errors == prev_num_errors or rerun_counter >= max_total_iterations:
            print(f"Stopping (no improvement or max iters). Overlaps left: {num_errors}")
            break

        # gather IDs to remove
        ids_to_remove = set()
        for err in overlaps:
            if not isinstance(err, str):
                continue
            parts = err.split("case_ids ")
            if len(parts) < 2:
                continue
            a, b = parts[1].split(" & ")
            id1, id2 = a.strip(), b.strip()
            src1 = final_schedule.loc[final_schedule['case_id']==id1, 'scheduled_by']
            src2 = final_schedule.loc[final_schedule['case_id']==id2, 'scheduled_by']
            if not src1.empty and not src2.empty:
                # drop CP if conflict
                ids_to_remove.add(id2 if src2.iloc[0]=='cp' else id1)

        removed = final_schedule[final_schedule['case_id'].isin(ids_to_remove)]
        final_schedule = final_schedule[~final_schedule['case_id'].isin(ids_to_remove)]
        for _, row in removed.iterrows():
            release_case_slots(
                row,
                avail_or_week, avail_surgeon_week,
                avail_anes_week, avail_team_week,
                buffer=buffer
            )

        to_reschedule = original_cases_df[
            original_cases_df['case_id'].astype(str).isin(ids_to_remove)
        ]
        if to_reschedule.empty:
            break

        if greedy_first:
            new_greedy, leftovers = greedy_fill_rooms_with_utilization_threshold(
                to_reschedule,
                avail_or_week, avail_surgeon_week,
                avail_anes=avail_anes_week,
                avail_team=avail_team_week,
                pref_sr=pref_sr,
                buffer=buffer
            )
            new_greedy['scheduled_by'] = 'greedy'
            to_cp = leftovers
        else:
            new_greedy = pd.DataFrame()
            to_cp = to_reschedule

        new_cp = pd.DataFrame()
        if not to_cp.empty:
            new_cp = cp_func(
                week_cases_df=to_cp,
                avail_or=avail_or_week,
                avail_surgeon=avail_surgeon_week,
                avail_anes=avail_anes_week,
                pref_sr=pref_sr,
                pref_sa=pref_sa,
                pref_st=pref_st,
                buffer=buffer,
                penalty_open_room=1
            )
            new_cp['scheduled_by'] = 'cp'

        final_schedule = pd.concat(
            [final_schedule, new_greedy, new_cp],
            ignore_index=True
        ).drop_duplicates(subset=['case_id'], keep='first')

        prev_num_errors = num_errors
        rerun_counter += 1

    return final_schedule

# --- Deep copy availability ---
avail_or_week    = copy.deepcopy(avail_or)
avail_surgeon_week = copy.deepcopy(avail_surgeon)
avail_anes_week  = copy.deepcopy(avail_anes_week)
avail_team_week  = copy.deepcopy(avail_team_week)

# --- Stage 1: Greedy Scheduling ---
print("Starting greedy scheduling...")
st = time.time()
greedy_df, leftovers_df = greedy_fill_rooms_with_utilization_threshold(
    busiest_week_cases,
    avail_or_week, avail_surgeon_week,
    avail_anes=avail_anes_week, avail_team=avail_team_week,
    pref_sr=pref_sr, buffer=buffer
)
greedy_df['scheduled_by'] = 'greedy'
if 'team' not in greedy_df.columns:
    greedy_df['team'] = greedy_df.get('team_all', [[]]*len(greedy_df))
print(f"Greedy scheduled {len(greedy_df)} | leftovers {len(leftovers_df)} in {round(time.time()-st,2)}s")
print("Checking overlaps after greedy:")
check_schedule_duplicates(greedy_df)

# --- Stage 2: Update availability ---
print("Updating availability after greedy...")
st2 = time.time()
for _, r in greedy_df.iterrows():
    room, day = r['room'], r['day_of_week']
    surgeon, anes = r['surgeon'], r.get('anes', None)
    team = r['team_all']
    rs, re = r['start_time'], r['end_time']+buffer
    avail_or_week[room][day]     = remove_time_slot(avail_or_week[room][day], rs, re)
    avail_surgeon_week[surgeon][day] = remove_time_slot(avail_surgeon_week[surgeon][day], rs, re)
    if pd.notnull(anes):
        avail_anes_week[anes][day] = remove_time_slot(avail_anes_week[anes][day], rs, re)
    for m in team:
        avail_team_week[m][day]   = remove_time_slot(avail_team_week[m][day], rs, re)
print(f"Availability updated in {round(time.time()-st2,2)}s")

# --- Stage 3: Constraint Programming (solve_week) for leftovers ---
print("Running CP for leftovers...")
cp_df = pd.DataFrame()
st3 = time.time()
if not leftovers_df.empty:
    cp_df = solve_week(
        week_cases_df=leftovers_df,
        avail_or=avail_or_week,
        avail_surgeon=avail_surgeon_week,
        avail_anes=avail_anes_week,
        pref_sr=pref_sr, pref_sa=pref_sa, pref_st=pref_st,
        buffer=buffer, penalty_open_room=1
    )
    cp_df['scheduled_by'] = 'cp'
    print(f"CP scheduled {len(cp_df)} in {round(time.time()-st3,2)}s")
else:
    print("No leftovers to CP.")
if 'team' not in cp_df.columns:
    cp_df['team'] = cp_df.get('team_all', [[]]*len(cp_df))
print("Checking overlaps after CP:")
check_schedule_duplicates(cp_df)

# --- Stage 4: Merge ---
print("Merging greedy + CP...")
greedy_df['case_id'] = greedy_df['case_id'].astype(str)
cp_df    ['case_id'] = cp_df   ['case_id'].astype(str)
already = set(greedy_df['case_id'])
cp_only = cp_df[~cp_df['case_id'].isin(already)]
final_schedule = pd.concat([greedy_df, cp_only], ignore_index=True).drop_duplicates(subset=['case_id'])
if 'team' not in final_schedule.columns:
    final_schedule['team'] = final_schedule.get('team_all', [[]]*len(final_schedule))
print(f"Merged total: {len(final_schedule)} cases")
print("Checking overlaps after merge:")
check_schedule_duplicates(final_schedule)

# --- Stage 5: Resolve overlaps & rerun CP ---
final_schedule = resolve_overlaps_and_rerun_cp(
    final_schedule, solve_week,
    avail_or_week, avail_surgeon_week,
    avail_anes_week, avail_team_week,
    pref_sr, pref_sa, pref_st,
    buffer, busiest_week_cases
)
print("Final schedule after overlap resolution:")
check_schedule_duplicates(final_schedule)

# --- Stage 6: Utilization Analysis ---
def calc_room_day_available(df, avail_surg):
    rows=[]
    for (room, day),grp in df.groupby(['room','day_of_week']):
        ints=[]
        for s in grp['surgeon'].unique():
            if day in avail_surg[s]:
                ints+=avail_surg[s][day]
        tot=sum(e-s for s,e in merge_intervals(ints))
        rows.append({'room':room,'day_of_week':day,'available':tot})
    return pd.DataFrame(rows)

def calc_room_day_util(df, avail_surg):
    used = df.assign(dur=lambda x: x['end_time']-x['start_time'])
    used = used.groupby(['room','day_of_week'])['dur'].sum().reset_index().rename(columns={'dur':'used'})
    avail = calc_room_day_available(df, avail_surg)
    m = pd.merge(used,avail,on=['room','day_of_week'],how='left')
    m['utilization'] = m['used'] / m['available']
    return m

print("Calculating utilization...")
ut = calc_room_day_util(final_schedule, avail_surgeon_week)
print(ut)
print("Top utilized:", ut.sort_values('utilization',ascending=False).head(5))
print("Least utilized:",ut.sort_values('utilization').head(5))
plt.figure(); ut['utilization'].hist(bins=20); plt.title('Utilization'); plt.show()

# --- Save results ---
final_schedule.to_csv('final_schedule.csv', index=False)
ut.to_csv('utilization.csv', index=False)
```

# **Rolling horizon**

### Monthly Scheduling Loop: Greedy + Constraint Programming (CP)

This script performs weekly scheduling for the least busy month in the dataset, using a hybrid approach:

1. **Preprocessing**: The dataset is enriched with `year`, `month`, `week_num`, and `day_of_week` columns.
2. **Selection**: Identifies the month with the fewest cases to test the scheduling pipeline.
3. **Deep Copies of Availability**: Creates isolated availability dictionaries to avoid mutating global state.
4. **Weekly Scheduling Loop**:
   - For each week:
     - First applies a greedy scheduling algorithm to assign as many cases as possible.
     - Then schedules the remaining (leftover) cases using the `solve_week` constraint programming model.
     - Releases assigned time slots back to availability pools after each assignment.
     - Adds metadata (`year`, `month`, `week_id`) and appends the result to a CSV file.

This routine provides a modular way to evaluate scheduling performance over a full month, while ensuring data isolation, resource tracking, and reproducibility.

```python
import os
import gc
import copy
import time
import pandas as pd

BUFFER       = 10      # Minutes buffer between cases
TIME_LIMIT   = 300     # Seconds time limit for CP
OUT_CSV      = 'final_schedule.csv'

def release_case_slots(case_row, avail_or, avail_surgeon, avail_anes, avail_team, buffer=BUFFER):
    room   = case_row['room']
    dow    = case_row['day_of_week']
    surg   = case_row['surgeon']
    anes   = case_row.get('anes', None)
    start0 = case_row['start_time']
    end0   = case_row['end_time']
    rs, re = start0, end0 + buffer

    if room in avail_or and dow in avail_or[room]:
        avail_or[room][dow].append((rs, re))
        avail_or[room][dow] = merge_intervals(avail_or[room][dow])

    if surg in avail_surgeon and dow in avail_surgeon[surg]:
        avail_surgeon[surg][dow].append((rs, re))
        avail_surgeon[surg][dow] = merge_intervals(avail_surgeon[surg][dow])

    if anes is not None and anes in avail_anes and dow in avail_anes[anes]:
        avail_anes[anes][dow].append((rs, re))
        avail_anes[anes][dow] = merge_intervals(avail_anes[anes][dow])

    for m in team:
        if m in avail_team and dow in avail_team[m]:
            avail_team[m][dow].append((rs, re))
            avail_team[m][dow] = merge_intervals(avail_team[m][dow])

def schedule_week(week_df, year, month, week_id,
                  avail_or, avail_surg, avail_anes, avail_team,
                  pref_sr, pref_sa, pref_st):
    print(f"  - scheduling week {week_id} ({len(week_df)} cases)")
    g_df, leftovers = greedy_fill_rooms_with_utilization_threshold(
        week_df,
        avail_or,
        avail_surg,
        avail_anes=avail_anes,
        avail_team=avail_team,
        pref_sr=pref_sr,
        buffer=BUFFER
    )
    g_df['scheduled_by'] = 'greedy'
    print(f"    greedy scheduled {len(g_df)}, leftovers {len(leftovers)}")
    for _, r in g_df.iterrows():
        release_case_slots(r, avail_or, avail_surg, avail_anes, avail_team, BUFFER)

    cp_df = pd.DataFrame()
    if not leftovers.empty:
        try:
            cp_df = solve_week(
            leftovers,
            avail_or,
            avail_surg,
            avail_anes,
            pref_sr,
            pref_sa,
            pref_st,
            BUFFER,
        )


            cp_df['scheduled_by'] = 'cp'
            print(f"    CP scheduled {len(cp_df)}")
            for _, r in cp_df.iterrows():
                release_case_slots(r, avail_or, avail_surg, avail_anes, avail_team, BUFFER)
        except Exception as e:
            print(f"    CP failed: {e}")
    else:
        print("    no leftovers → skip CP")

    out = pd.concat([g_df, cp_df], ignore_index=True).drop_duplicates('case_id')
    out['year']    = year
    out['month']   = month
    out['week_id'] = week_id

    header = not os.path.exists(OUT_CSV)
    out.to_csv(OUT_CSV, mode='a', header=header, index=False)
    return out

# --------------------------------------------------------------------------------
# Main Loop
# --------------------------------------------------------------------------------
new_cases_df['date']         = pd.to_datetime(new_cases_df['date'])
new_cases_df['year']         = new_cases_df['date'].dt.year
new_cases_df['month']        = new_cases_df['date'].dt.month
new_cases_df['day_of_week']  = new_cases_df['date'].dt.weekday
new_cases_df['week_num']     = new_cases_df['date'].dt.isocalendar().week

counts = new_cases_df.groupby(['year','month']).size()
least_year, least_month = counts.idxmin()
print(f"Least-busy month: {least_year}-{least_month} ({counts.min()} cases)")

month_df = new_cases_df[
    (new_cases_df['year']==least_year) &
    (new_cases_df['month']==least_month)
].copy()

or_av    = copy.deepcopy(avail_or)
surg_av  = copy.deepcopy(avail_surgeon)
anes_av  = copy.deepcopy(avail_anes_week)
team_av  = copy.deepcopy(avail_team_week)

if os.path.exists(OUT_CSV):
    os.remove(OUT_CSV)

for wk in sorted(month_df['week_num'].unique()):
    week_df = month_df[month_df['week_num']==wk]
    if week_df.empty:
        continue
    schedule_week(
        week_df, least_year, least_month, wk,
        or_av, surg_av, anes_av, team_av,
        pref_sr, pref_sa, pref_st
    )
    gc.collect()
```
